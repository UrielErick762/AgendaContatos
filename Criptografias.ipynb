Fernet (criptografia simétrica) usando um módulo de criptografia em Python


A codificação simétrica é uma técnica criptográfica, onde a mesma chave é usada para criptografar e descriptografar mensagens do cliente para o servidor. Para garantir que nenhuma informação confidencial seja vazada durante a passagem dos pacotes de rede por servidores vulneráveis, onde os hackers podem usar esta mensagem para fins maliciosos, criptografar a mensagem será uma boa ideia.

```python
from cryptography.fernet import Fernet


```

Fernet é um módulo de criptografia simétrica autenticada incluído na biblioteca cryptography do Python, projetado para garantir que mensagens não possam ser lidas ou manipuladas sem a chave secreta correta.  Ele utiliza AES em modo CBC com padding PKCS7 para criptografia e HMAC com SHA256 para autenticação, gerando tokens seguros que incluem um carimbo de tempo e um vetor de inicialização aleatório.

```python
chave = Fernet.generate_key()
fernet = Fernet(chave)

mensagem = input("digite uma mensagem para criptografar:")

msg_bytes = mensagem.encode()
encrypted_msg = fernet.encrypt(msg_bytes)
decrypted_msg = fernet.decrypt(encrypted_msg)
decrypted_msg = decrypted_msg.decode()

```

```python
print("Mensagem original: ", decrypted_msg)
print("Mensagem em bytes:", list(msg_bytes))
print("Mensagem criptografada: ", encrypted_msg)
print("Mensagem descriptografada: ", decrypted_msg)
```
